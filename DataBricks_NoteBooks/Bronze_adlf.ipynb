##For the bulk rides

```python
# The SAS URL for the historical bulk data
url = "https://dluberprojectdev.blob.core.windows.net/githubdata/data/bulk_rides.json.json?sp=r&st=2026-04-17T06:48:43Z&se=2026-04-20T15:03:43Z&spr=https&sv=2025-11-05&sr=c&sig=4AY6179KydZFtUwvK%2BdNCrvjfrcopCxfav9ExSPdjkg%3D"

# Reading the data using Pandas
df = pd.read_json(url)

# Converting Pandas DataFrame to Spark DataFrame
df_spark = spark.createDataFrame(df)

# Check if table exists; if not, perform the one-time bulk write
if not spark.catalog.tableExists("uber_project.bronze_layer.bulk_rides"):
    df_spark.write.format("delta") \
        .mode("overwrite") \
        .saveAsTable("uber_project.bronze_layer.bulk_rides")
    
    print("This will not run more than 1 time")
```
