## Streaming rides Transformation

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%sql
select * from uber_project.bronze_layer.rides_raw
```

```python
df=spark.read.table("uber_project.bronze_layer.bulk_rides")
df_rides_schema=df.schema
display(df.schema) 
```

```python
df_rides=spark.read.table("uber_project.bronze_layer.rides_raw")

df_parsed=df_rides.withColumn("parsed_rides",from_json(col("rides"), df_rides_schema)).select("parsed_rides.*")
```

```python
display(df_parsed)
```
